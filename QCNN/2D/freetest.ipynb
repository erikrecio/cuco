```python
X_train, X_test, Y_train, Y_test = data2D.data_load_and_process('sinus2')
X_train.shape, X_test.shape, len(Y_train), len(Y_test)
```

```python
import sklearn.svm as svm

SVM = svm.SVC(kernel='poly', degree=2)
#SVM  = svm.SVC(kernel='rbf', C=10)
#SVM  = svm.SVC(kernel='sigmoid')
SVM.fit(X_train, Y_train);
```

```python
preds = SVM.predict(X_train)
trues = np.array(Y_train)
comp = preds==trues

fig, ax = plt.subplots(1,2, figsize=(8,10))

ax[0].set_aspect('equal')
ax[0].scatter(X_train.T[0], X_train.T[1], s=10, cmap='rainbow', c=trues);
ax[0].set_title('true train')
ax[1].set_aspect('equal')
ax[1].scatter(X_train.T[0], X_train.T[1], s=10, cmap='rainbow', c=preds*comp);
ax[1].set_title('preds train');

print('Train-set accuracy:',np.sum(preds==trues)/len(trues))
```

```python
preds = SVM.predict(X_test)
trues = np.array(Y_test)

comp = preds==trues

fig, ax = plt.subplots(1,2, figsize=(8,10))

ax[0].set_aspect('equal')
ax[0].scatter(X_test.T[0], X_test.T[1], s=10, cmap='rainbow', c=trues);
ax[0].set_title('true train')
ax[1].set_aspect('equal')
ax[1].scatter(X_test.T[0], X_test.T[1], s=10, cmap='rainbow', c=preds*comp);
ax[1].set_title('preds train');
print('Test-set accuracy:',np.sum(preds==trues)/len(trues))
```
